## 1. Introduction to the problem of study [1,0 valor]

### Problem Context

In the insurance industry, determining appropriate compensation for vehicle accident claims is a complex decision-making process that involves analyzing multiple uncertain and imprecise factors. Traditional binary logic systems often fail to capture the nuanced reality of accident scenarios, where variables like damage severity and driver fault exist on a continuous spectrum rather than in discrete categories.

### The Challenge

Insurance adjusters must evaluate:
- *Damage Severity*: The extent of vehicle damage (ranging from minor scratches to total loss)
- *Driver Fault*: The degree of responsibility attributed to the insured driver (from completely innocent to entirely at fault)
- *Witness Testimony*: Additional evidence that may support or contradict the initial assessment

The relationship between these variables is not straightforward. For instance, high fault typically reduces payment, but the interaction with damage severity creates complex scenarios:
- High fault + High damage â†’ Low payment (driver primarily responsible)
- Low fault + High damage â†’ High payment (significant compensation needed)
- Low fault + Low damage â†’ Low payment (minor incident, minor compensation)
- Moderate fault + Moderate damage â†’ Uncertain (requires fuzzy reasoning)

### Why Fuzzy Logic?

Classical binary logic cannot adequately model linguistic terms like "slightly damaged," "moderately at fault," or "somewhat credible witnesses." Fuzzy logic provides a mathematical framework to handle this imprecision, allowing the system to:

1. Model human reasoning and expert judgment
2. Handle uncertainty and vagueness in input data
3. Provide smooth, continuous transitions between decision boundaries
4. Incorporate multiple variables with complex interdependencies

This project implements a *Fuzzy Inference System (FIS)* to automate and standardize insurance payment decisions while maintaining the flexibility to handle real-worldÂ ambiguity.

## 2. Brief and general description of the approach and methods used [1,5 valor]

### Overall Approach

This project employs a *Fuzzy Inference System (FIS)* based on the Mamdani method to model the insurance claim decision-making process. The system transforms crisp numerical inputs into linguistic variables, applies expert-defined rules, and produces a defuzzified payment recommendation.

### System Architecture

The fuzzy logic system consists of four main components:

#### 1. *Fuzzification*
Converts crisp input values into degrees of membership across multiple fuzzy sets:
- *Damage Input*: Mapped to linguistic terms (e.g., "Low," "Medium," "High")
- *Fault Input*: Categorized by responsibility level (e.g., "None," "Partial," "Complete")
- *Witness Input*: Assessed for credibility and relevance (e.g., "Weak," "Moderate," "Strong")

Each input variable uses membership functions (typically triangular or trapezoidal) to determine the degree to which the input belongs to each fuzzy set.

#### 2. *Rule Base*
A collection of IF-THEN rules that encode expert knowledge and insurance policies:

- IF damage IS high AND fault IS low THEN payment IS high
- IF damage IS low AND fault IS high THEN payment IS very_low
- IF damage IS medium AND fault IS medium AND witnesses IS strong THEN payment IS medium_high


These rules capture the complex interactions between variables that would be difficult to express in traditional mathematical formulas.

#### 3. *Inference Engine*
Applies fuzzy logic operators to evaluate all rules simultaneously:
- *AND operator*: Typically implemented as minimum (min) or product
- *OR operator*: Typically implemented as maximum (max)
- *Aggregation*: Combines the outputs of all activated rules using max or sum

The engine determines which rules fire and to what degree based on the fuzzified inputs.

#### 4. *Defuzzification*
Converts the fuzzy output into a crisp payment value using methods such as:
- *Centroid Method*: Calculates the center of gravity of the aggregated fuzzy output
- *Mean of Maximum*: Takes the average of the values with maximum membership
- *Weighted Average*: Computes a weighted mean based on rule strengths

### Methodology Steps

1. *Define Input/Output Variables*: Identify linguistic variables and their ranges
2. *Design Membership Functions*: Create appropriate shapes (triangular, Gaussian, etc.)
3. *Establish Rule Base*: Formulate expert rules based on domain knowledge
4. *Configure Inference System*: Select operators and aggregation methods
5. *Implement in Python*: Use libraries like scikit-fuzzy or simpful
6. *Validate and Test*: Compare outputs against expert decisions and edge cases
7. *Optimize*: Tune membership functions and rules to improve accuracy

### Advantages of This Approach

- *Handles Uncertainty*: Manages imprecise and incomplete information naturally
- *Interpretable*: Rules are human-readable and align with expert reasoning
- *Flexible*: Easy to modify rules or add new variables without restructuring the entire system
- *Non-linear Modeling*: Captures complex relationships without requiring explicit mathematical formulations
- *Robust*: Performs well even with noisy or missing data

### Expected Outcomes

The system will produce payment recommendations that:
- Reflect realistic expert judgment
- Provide consistent decisions for similar cases
- Handle edge cases and ambiguous scenarios gracefully
- Offer transparency through ruleÂ traceability

## 3. Brief History and literature review of the problem and methods/algorithms [1,5 valor]

### 3.1 Historical Context of Vehicle Crash Analysis

The analysis of vehicle crashes and their implications for insurance has been a subject of research for several decades. As road traffic accidents became a leading cause of mortality and economic loss globally, the need for accurate risk assessment and claim prediction systems grew significantly. The World Health Organization (WHO) reported that road traffic deaths reached 1.35 million annually, making it the leading killer of people aged 5-29 years [1].

Traditional approaches to insurance claim assessment relied heavily on deterministic models and crisp logic systems. However, these methods often failed to capture the inherent uncertainty and imprecision in accident scenarios, where factors like damage severity, driver fault, and witness credibility exist on continuous spectrums rather than in discrete categories.

### 3.2 Evolution of Fuzzy Logic in Insurance

The application of fuzzy logic to insurance problems was pioneered by DeWit in 1982 [5], who first recognized that underwriting decisions involved subjective assessments that could not be adequately captured by binary logic [2]. This seminal work laid the foundation for incorporating linguistic variables and fuzzy set theory into insurance decision-making processes.

Over the past 25 years, the use of fuzzy logic in insurance has expanded considerably [2]. FL technologies have been employed in insurance-related areas including:

- **Classification and Underwriting**: Lemaire (1990) used fuzzy expert systems to provide flexible definitions of preferred policyholders, employing continuous membership functions and various fuzzy intersection operators [2].

- **Risk Assessment**: Multiple studies have demonstrated that fuzzy set theory provides a realistic approach to formal risk analysis, particularly when dealing with imprecise definitions and measurements [2].

- **Pricing and Ratemaking**: Young (1996, 1997) [6] illustrated how fuzzy logic could be used to make pricing decisions in group health insurance that consistently consider supplementary data, including vague or linguistic objectives of the insurer [2].

### 3.3 Machine Learning Approaches to Driver Behavior Analysis

Recent advances in sensor technology and machine learning have enabled more sophisticated approaches to driver behavior analysis. Yuksel & Atmaca [1] developed a comprehensive driver risk assessment system using machine learning and fuzzy logic, achieving 100% accuracy in identifying risky driving behaviors.

Their study compared seven widely-used machine learning algorithms:
- C4.5 Decision Tree (98.37% accuracy)
- Random Forest (99.82% accuracy)
- Artificial Neural Network (99.73% accuracy)
- Support-Vector Machine (98.37% accuracy)
- K-Nearest Neighbor (99.91% accuracy)
- Naive Bayes (93.01% accuracy)
- **K-Star (100% accuracy)** - the best performing algorithm

The authors demonstrated that using accelerometer and gyroscope sensors with a rolling window approach and statistical features could effectively identify four major risky driving behaviors: sudden acceleration, sudden deceleration, sudden right turn, and sudden left turn [1].

### 3.4 Vehicle Crash Modeling Using Neural-Fuzzy Systems

The complexity of vehicle crash dynamics has led researchers to explore hybrid approaches combining multiple computational intelligence techniques. Zhao et al. [3] presented a novel Adaptive Neural-Fuzzy Inference System (ANFIS) approach to reconstruct kinematics of colliding vehicles.

### 3.5 Risk Factor Prioritization in Insurance

Understanding which factors contribute most significantly to risk is crucial for fair and accurate insurance pricing. Esfandabadi et al. [4] addressed this challenge in the context of comprehensive automobile insurance in Iran using a hybrid multi-criteria decision-making model combining:

- **Fuzzy Delphi Method (FDM)**: To identify important risk factors through expert consensus
- **Fuzzy Analytic Hierarchy Process (FAHP)**: To prioritize and weight the identified factors
- **Similarity Aggregation Method (SAM)**: To combine individual expert opinions into group consensus

Their findings revealed that behavioral characteristics and driving experience were more important than vehicle specifications in determining risk levels. The top three risk factors identified were [4]:
1. Traffic offences including speed violations (highest weight)
2. Claim history in third-party liability insurance
3. Driving experience

## 7. Main code as possible solution to the problem [1,5 valor] 

```python
# Function to calculate Fault Score (0-100)
def calculate_fault_score(row):
    """
    Calculate fault score based on:
    - Primary Factor: Different factors have different fault weights
    - Hour: Late night/early morning hours increase fault
    - Weekend: Weekend incidents may indicate riskier behavior
    """
    score = 0
    
    # Primary Factor contribution (0-60 points)
    primary_factor = str(row['Primary Factor']).upper()
    
    # Highest fault - Reckless behavior (60 points)
    if any(keyword in primary_factor for keyword in [
        'UNSAFE SPEED', 'SPEED TOO FAST', 'ALCOHOLIC BEVERAGES', 
        'ILLEGAL DRUGS', 'DRIVER ASLEEP', 'DRIVER DISTRACTED',
        'CELL PHONE', 'RECKLESS', 'WRONG WAY'
    ]):
        score += 60
    
    # High fault - Violations (50 points)
    elif any(keyword in primary_factor for keyword in [
        'DISREGARD SIGNAL', 'RAN OFF ROAD', 'LEFT OF CENTER',
        'IMPROPER PASSING', 'IMPROPER LANE USAGE', 'UNSAFE LANE MOVEMENT'
    ]):
        score += 50
    
    # Medium-high fault - Failure to act properly (40 points)
    elif any(keyword in primary_factor for keyword in [
        'FAILURE TO YIELD', 'FOLLOWING TOO CLOSELY', 'IMPROPER TURNING',
        'UNSAFE BACKING', 'OVERCORRECTING', 'OVERSTEERING'
    ]):
        score += 40
    
    # Medium fault - Driver condition or distraction (35 points)
    elif any(keyword in primary_factor for keyword in [
        'DRIVER ILLNESS', 'PRESCRIPTION DRUGS', 'FATIGUED',
        'PASSENGER DISTRACTION', 'VIEW OBSTRUCTED', 'GLARE'
    ]):
        score += 35
    
    # Low-medium fault - Vehicle defects (25 points)
    elif any(keyword in primary_factor for keyword in [
        'BRAKE FAILURE', 'DEFECTIVE', 'TIRE FAILURE', 
        'STEERING FAILURE', 'ENGINE FAILURE', 'HEADLIGHT'
    ]):
        score += 25
    
    # Low fault - Environmental factors (15 points)
    elif any(keyword in primary_factor for keyword in [
        'ROADWAY SURFACE', 'ANIMAL/OBJECT', 'WEATHER CONDITIONS',
        'CROSSWINDS', 'OBSTRUCTION', 'ROAD UNDER CONSTRUCTION'
    ]):
        score += 15
    
    # Minimal fault - No driver fault (5 points)
    elif 'NONE' in primary_factor:
        score += 5
    
    # Default for other factors (30 points)
    else:
        score += 30
    
    # Hour contribution (0-20 points) - risky hours increase fault
    # Note: Hour is in military time format (e.g., 1700 = 5:00 PM)
    hour = row['Hour']
    if pd.notna(hour):
        hour_int = int(hour) // 100  # Convert 1700 to 17
        if (0 <= hour_int < 6) or (22 <= hour_int < 24):  # Late night/early morning
            score += 15
        elif 16 <= hour_int < 20:  # Rush hour/evening
            score += 10
        else:
            score += 5
    else:
        score += 5  # Default if hour is missing
    
    # Weekend contribution (0-20 points)
    if row['Weekend?'] == 'Weekend':
        score += 15
    else:
        score += 5
    
    return min(score, 100)  # Cap at 100


# Function to calculate Severity Score (0-100)
def calculate_severity_score(row):
    """
    Calculate severity score based on:
    - Collision Type: More severe collision types score higher
    - Injury Type: More serious injuries score higher
    """
    score = 0
    
    # Collision Type contribution (0-50 points)
    collision_type = str(row['Collision Type']).upper()
    
    if 'PEDESTRIAN' in collision_type:
        score += 50  # Pedestrian collisions are very severe
    elif 'CYCLIST' in collision_type or 'MOPED' in collision_type or 'MOTORCYCLE' in collision_type:
        score += 45  # Vulnerable road users
    elif '3+' in collision_type or 'BUS' in collision_type:
        score += 40  # Multi-vehicle or bus collisions
    elif '2-CAR' in collision_type or '2 CAR' in collision_type:
        score += 30  # Standard two-car collision
    elif '1-CAR' in collision_type or '1 CAR' in collision_type:
        score += 20  # Single vehicle incident
    else:
        score += 25  # Default for unspecified
    
    # Injury Type contribution (0-50 points)
    injury_type = str(row['Injury Type']).upper()
    
    if 'FATAL' in injury_type:
        score += 50  # Maximum severity for fatalities
    elif 'INCAPACITATING' in injury_type:
        score += 40  # Serious injuries requiring hospitalization
    elif 'NON-INCAPACITATING' in injury_type:
        score += 25  # Minor injuries
    elif 'NO INJURY' in injury_type or 'UNKNOWN' in injury_type:
        score += 5   # Property damage only or unknown
    else:
        score += 15  # Default for other cases
    
    return min(score, 100)  # Cap at 100
```

```python
fault = ctrl.Antecedent(np.arange(0, 101, 1), 'fault')
severity = ctrl.Antecedent(np.arange(0, 101, 1), 'severity')
payment = ctrl.Consequent(np.arange(0, 101, 1), 'payment')

fault['low'] = fuzz.trimf(fault.universe, [0, 0, 40])
fault['medium'] = fuzz.trimf(fault.universe, [20, 50, 80])
fault['high'] = fuzz.trimf(fault.universe, [60, 100, 100])

severity['low'] = fuzz.trimf(severity.universe, [0, 0, 40])
severity['medium'] = fuzz.trimf(severity.universe, [20, 50, 80])
severity['high'] = fuzz.trimf(severity.universe, [60, 100, 100])

payment['verylow'] = fuzz.trimf(payment.universe, [0, 0, 25])
payment['low'] = fuzz.trimf(payment.universe, [10, 30, 50])
payment['medium'] = fuzz.trimf(payment.universe, [40, 60, 75])
payment['high'] = fuzz.trimf(payment.universe, [70, 100, 100])

rule1 = ctrl.Rule(severity['high'] & fault['low'], payment['high'])
rule2 = ctrl.Rule(severity['high'] & fault['medium'], payment['medium'])
rule3 = ctrl.Rule(severity['high'] & fault['high'], payment['low'])
rule4 = ctrl.Rule(severity['medium'] & fault['low'], payment['medium'])
rule5 = ctrl.Rule(severity['medium'] & fault['medium'], payment['medium'])
rule6 = ctrl.Rule(severity['medium'] & fault['high'], payment['low'])
rule7 = ctrl.Rule(severity['low'] & fault['low'], payment['low'])
rule8 = ctrl.Rule(severity['low'] & fault['medium'], payment['verylow'])
rule9 = ctrl.Rule(severity['low'] & fault['high'], payment['verylow'])

payment_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
payment_sim = ctrl.ControlSystemSimulation(payment_ctrl)

def calcula_payment(fault_score, severity_score):
    payment_sim.input['fault'] = fault_score
    payment_sim.input['severity'] = severity_score
    payment_sim.compute()
    return payment_sim.output['payment']
```

```python
def visualize_fuzzy_inference(fault_val, severity_val):
    
    payment_val = calcula_payment(fault_val, severity_val)
    
    fig = plt.figure(figsize=(16, 10))
    fig.suptitle(f'Fuzzy Inference Process - Insurance Payment\nFault Score = {fault_val} | Severity Score = {severity_val} | Payment = {payment_val:.2f}%', 
                 fontsize=16, fontweight='bold')
    
    # ==================== INPUT 1: FAULT ====================
    ax1 = plt.subplot(2, 2, 1)
    x = fault.universe
    ax1.plot(x, fuzz.trimf(x, [0, 0, 40]), 'b', linewidth=2, label='Low')
    ax1.plot(x, fuzz.trimf(x, [20, 50, 80]), 'orange', linewidth=2, label='Medium')
    ax1.plot(x, fuzz.trimf(x, [60, 100, 100]), 'g', linewidth=2, label='High')
    
    ax1.axvline(x=fault_val, color='black', linestyle='--', linewidth=2.5)
    
    fault_low_deg = fuzz.interp_membership(x, fuzz.trimf(x, [0, 0, 40]), fault_val)
    fault_med_deg = fuzz.interp_membership(x, fuzz.trimf(x, [20, 50, 80]), fault_val)
    fault_high_deg = fuzz.interp_membership(x, fuzz.trimf(x, [60, 100, 100]), fault_val)
    
    ax1.plot(fault_val, fault_low_deg, 'bo', markersize=12)
    ax1.plot(fault_val, fault_med_deg, 'o', color='orange', markersize=12)
    ax1.plot(fault_val, fault_high_deg, 'go', markersize=12)
    
    ax1.text(0.02, 0.98, f'Input = {fault_val}', transform=ax1.transAxes,
             fontsize=11, verticalalignment='top', fontweight='bold',
             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
    
    ax1.set_xlabel('Fault Score', fontsize=12, fontweight='bold')
    ax1.set_ylabel('Membership', fontsize=12, fontweight='bold')
    ax1.set_title('Input: Fault', fontsize=14, fontweight='bold')
    ax1.legend(loc='upper right')
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim([-0.05, 1.05])
    

    # ==================== INPUT 2: SEVERITY ====================
    ax2 = plt.subplot(2, 2, 2)
    x = severity.universe
    ax2.plot(x, fuzz.trimf(x, [0, 0, 40]), 'b', linewidth=2, label='Low')
    ax2.plot(x, fuzz.trimf(x, [20, 50, 80]), 'orange', linewidth=2, label='Medium')
    ax2.plot(x, fuzz.trimf(x, [60, 100, 100]), 'g', linewidth=2, label='High')
    
    ax2.axvline(x=severity_val, color='black', linestyle='--', linewidth=2.5)
    
    sev_low_deg = fuzz.interp_membership(x, fuzz.trimf(x, [0, 0, 40]), severity_val)
    sev_med_deg = fuzz.interp_membership(x, fuzz.trimf(x, [20, 50, 80]), severity_val)
    sev_high_deg = fuzz.interp_membership(x, fuzz.trimf(x, [60, 100, 100]), severity_val)
    
    ax2.plot(severity_val, sev_low_deg, 'bo', markersize=12)
    ax2.plot(severity_val, sev_med_deg, 'o', color='orange', markersize=12)
    ax2.plot(severity_val, sev_high_deg, 'go', markersize=12)
    
    ax2.text(0.02, 0.98, f'Input = {severity_val}', transform=ax2.transAxes,
             fontsize=11, verticalalignment='top', fontweight='bold',
             bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))
    
    ax2.set_xlabel('Severity Score', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Membership', fontsize=12, fontweight='bold')
    ax2.set_title('Input: Severity', fontsize=14, fontweight='bold')
    ax2.legend(loc='upper right')
    ax2.grid(True, alpha=0.3)
    ax2.set_ylim([-0.05, 1.05])
    
    
    # ==================== OUTPUT: PAYMENT ====================
    ax3 = plt.subplot(2, 1, 2)
    x = payment.universe
    
    ax3.plot(x, fuzz.trimf(x, [0, 0, 25]), 'purple', linewidth=2, alpha=0.7, label='Very Low')
    ax3.plot(x, fuzz.trimf(x, [10, 30, 50]), 'b', linewidth=2, alpha=0.7, label='Low')
    ax3.plot(x, fuzz.trimf(x, [40, 60, 75]), 'orange', linewidth=2, alpha=0.7, label='Medium')
    ax3.plot(x, fuzz.trimf(x, [70, 100, 100]), 'g', linewidth=2, alpha=0.7, label='High')
    
    try:
        payment.view(sim=payment_sim, ax=ax3)
    except:
        pass
    
    ax3.axvline(x=payment_val, color='red', linestyle='--', linewidth=4, alpha=0.8,
                label=f'Defuzzified Output = {payment_val:.2f}%')
    
    ax3.plot(payment_val, 0, 'ro', markersize=15, zorder=5)
    
    ax3.text(payment_val, 1.02, f'{payment_val:.1f}%', 
             ha='center', fontsize=12, fontweight='bold',
             bbox=dict(boxstyle='round', facecolor='red', alpha=0.7, edgecolor='black'))
    
    ax3.set_xlabel('Payment (%)', fontsize=12, fontweight='bold')
    ax3.set_ylabel('Membership', fontsize=12, fontweight='bold')
    ax3.set_title('Output: Insurance Payment (Aggregation & Defuzzification)', 
                  fontsize=14, fontweight='bold')
    ax3.legend(loc='upper left', fontsize=10)
    ax3.grid(True, alpha=0.3)
    ax3.set_ylim([-0.05, 1.1])
    
    plt.subplots_adjust(hspace=0.4)
    plt.show()
    
    # Print summary
    print("\n" + "="*70)
    print("FUZZY INFERENCE SUMMARY")
    print("="*70)
    print(f"\nINPUTS:")
    print(f"   Fault Score:    {fault_val:6.2f}")
    print(f"   Severity Score: {severity_val:6.2f}")
    
    print(f"\nMEMBERSHIP DEGREES:")
    print(f"   Fault:    Low={fault_low_deg:.3f} | Medium={fault_med_deg:.3f} | High={fault_high_deg:.3f}")
    print(f"   Severity: Low={sev_low_deg:.3f} | Medium={sev_med_deg:.3f} | High={sev_high_deg:.3f}")
    
    print(f"\nOUTPUT:")
    print(f"   Insurance Payment: {payment_val:.2f}%")
    print("="*70 + "\n")
```

## 10. Pros and cons of the approach [2,0 valor]


### ADVANTAGES

#### 1. Handling of Uncertainty and Imprecision
Traditional binary logic systems struggle with the inherent ambiguity in accident scenarios.  
Fuzzy logic naturally accommodates linguistic terms like *"moderate fault"* or *"high severity"* without forcing artificial binary classifications.  
This aligns with how human experts naturally reason about insurance claims.

#### 2. Interpretability and Transparency
The rule-based structure provides clear traceability of decisions.  
Each rule represents an explicit policy that can be audited, explained to stakeholders, and modified if necessary.  
This is crucial for regulatory compliance and customer trust in insurance contexts.

#### 3. Smooth Output Transitions
Unlike threshold-based systems that produce abrupt changes in recommendations, fuzzy logic generates gradual transitions.  
Small changes in input scores result in proportional output adjustments, avoiding arbitrary discontinuities.

#### 4. Expert Knowledge Integration
The system could directly encode domain expertise through membership functions and rules.  
Insurance adjusters' experience and industry standards can be formalized without large datasets or complex ML models.

#### 5. Computational Efficiency
Fuzzy inference systems are computationally lightweight compared to deep learning alternatives.  
Real-time claim processing is feasible even with limited hardware.

#### 6. Robustness to Incomplete Data
The system continues to function when some input features are missing or uncertain.  
Fuzzy membership functions naturally handle boundary cases and noisy measurements.

---

### LIMITATIONS

#### 1. Rule Complexity and Maintenance
As the number of input variables grows, the rule base expands exponentially.  
Our system with 2 inputs and 3 functions needs only 9 rules, but adding a third variable would require 27 rules.

#### 2. Membership Function Tuning
Defining appropriate membership functions requires significant domain expertise and iterative refinement.  
Triangular functions are simple but may not capture true underlying distributions.

#### 3. Limited Learning Capability
Unlike machine learning, the fuzzy system does not automatically adapt to new data.  
Rules and membership functions must be manually updated.

#### 4. Contextual Blindness
The system cannot distinguish between scenarios that produce identical input scores.  
Critical contextual information such as right-of-way or driver impairment is lost during scoring.

#### 5. Oversimplified Feature Engineering
Aggregating multiple factors into single scores loses nuance.  
A *fault score* of 70 from speeding is treated the same as one from distracted driving.

#### 6. No Confidence Estimation
The system outputs a payment percentage without indicating confidence.  
Ambiguous cases should flag for human review, but current implementation treats all as equally reliable.

#### 7. Static Policy Representation
Insurance policies vary by jurisdiction and type.  
A fixed rule set cannot adapt without creating separate systems for each context.


### CONCLUSION

The fuzzy logic approach provides a practical balance between interpretability and performance for insurance claim assessment.  
Its primary strengths lie in transparency, expert knowledge integration, and natural uncertainty handling.  

However, it is fundamentally limited by its inability to learn from data and its dependence on contextual information.  
For production deployment, a hybrid system combining fuzzy inference for routine cases with human review for ambiguous scenarios would be most appropriate.
